```python
#randomwalk in 1000 steps
import random
import numpy as np
import matplotlib.pyplot as plt
position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)

nsteps = 1000
draws = np.random.randint(0, 2, size = nsteps)
steps = np.where( draws > 0, 1, -1)
walk = steps.cumsum()
plt.plot(walk)
plt.show()
```
