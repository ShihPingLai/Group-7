```python
#TwoDimensionalRandomWalk

import random
import numpy as np
import matplotlib.pyplot as plt
import math


frequency = 1000
displacementx = [0]
displacementy = [0]
mux = 0
muy = 0
sigmax = 2
sigmay = 2

positionx = 0
positiony = 0
amplitudex = 20
amplitudey = 20


for i in range(frequency):
    probabilityx = np.random.normal(mux, sigmay, size = None)
    positionx += probabilityx * amplitudex
    displacementx.append(positionx)
    
    probabilityy = np.random.normal(mux, sigmay, size = None)
    positiony += probabilityy * amplitudey
    displacementy.append(positiony)

    
plt.plot(displacementx,displacementy)
plt.show()

```

```python
#OneDimensionalRandomWalk

import random
import numpy as np
import matplotlib.pyplot as plt
import math


frequency = 1000
displacement = [0]
mu = 0
sigma = 2

position = 0
amplitude = 20



for i in range(frequency):
    probability = np.random.normal(mu, sigma, size = None)
    position += probability * amplitude
    displacement.append(position)
    

    
plt.plot(displacement)
plt.show()
```
