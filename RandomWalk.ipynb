```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math


frequency = 1000
displacementx = [0]
displacementy = [0]
mux = 0
muy = 0
sigmax = 1
sigmay = 1

positionx = 0
positiony = 0
amplitudex = 20
amplitudey = 20


for i in range(frequency):
    x = np.random.normal(mux, sigmay, size = None)
    positionx += x * amplitudex
    displacementx.append(positionx)
    
    y = np.random.normal(mux, sigmay, size = None)
    positiony += y * amplitudey
    displacementy.append(positiony)

    
plt.plot(displacementx,displacementy)
plt.show()

```
