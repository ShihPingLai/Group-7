```python
#randomwalk in 10000 steps
import random
import numpy as np
import matplotlib.pyplot as plt


positionx = 0
walkx = [positionx]
steps = 10000
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    positionx += step
    walkx.append(positionx)

nsteps = 10000
draws = np.random.randint(0, 2, size = nsteps)
steps = np.where( draws > 0, 1, -1)
walkx = steps.cumsum()

positiony = 0
walky = [positiony]
steps = 10000
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    positiony += step
    walky.append(positiony)

nsteps = 10000
draws = np.random.randint(0, 2, size = nsteps)
steps = np.where( draws > 0, 1, -1)
walky = steps.cumsum()


plt.plot(walkx,walky)
plt.show()
```
